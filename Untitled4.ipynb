```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
df1 = pd.read_csv('train.csv')
df2 = pd.read_csv('test.csv')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.isnull().sum()
```

```python
df1['Arrival Delay in Minutes']=df1['Arrival Delay in Minutes'].fillna(df1['Arrival Delay in Minutes'].median())
```

```python
df1.isnull().sum()
```

```python
df1['Gender']=df1['Gender'].map({'Male': 1, 'Female': 0})
df1['Customer Type']=df1['Customer Type'].map({'Loyal Customer': 1, 'disloyal Customer': 0})
df1['Type of Travel']=df1['Type of Travel'].map({'Business travel': 1, 'Personal Travel': 0})
```

```python
df1 = pd.get_dummies(df1, columns=['Class'])
```

```python
df1= df1.apply(lambda col: col.astype(int) if col.dtype == "bool" else col)
```

```python
X= df1.drop(['satisfaction','Gender','Id','Age','Type of Travel','Flight Distance','Seat comfort','Departure/Arrival time convenient','Food and drink','Gate location','Inflight wifi service','Departure Delay in Minutes','Arrival Delay in Minutes','Class_Eco Plus'],axis=1)
y = df1['satisfaction']
```

```python
tree=DecisionTreeClassifier(random_state=42)
tree.fit(X,y)
```

```python
df2['Gender']=df2['Gender'].map({'Male': 1, 'Female': 0})
df2['Customer Type']=df2['Customer Type'].map({'Loyal Customer': 1, 'disloyal Customer': 0})
df2['Type of Travel']=df2['Type of Travel'].map({'Business travel': 1, 'Personal Travel': 0})
```

```python
df2 = pd.get_dummies(df2, columns=['Class'])
```

```python
df2= df2.apply(lambda col: col.astype(int) if col.dtype == "bool" else col)
```

```python
X_test=df2.copy()
```

```python
X_test=X_test.drop(['Gender','Id','Age','Type of Travel','Flight Distance','Seat comfort',
                    'Departure/Arrival time convenient','Food and drink','Gate location',
                    'Inflight wifi service','Departure Delay in Minutes','Arrival Delay in Minutes','Class_Eco Plus'],axis=1)
```

```python
prediction=tree.predict(X_test)
```

```python
prediction
```

```python
submission= pd.DataFrame({'ID' : df2['Id'], 'satisfatction' : prediction})
```
