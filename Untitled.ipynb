#### As we can see the only column that has any null value is "Arrival Delay in Minutes"
Here we can handle these null values by either mean or median. Because of long delays in some cases this column has some extreme values so we will be using median here as mean is affected much more by values compared to median.

```python
df1['Arrival Delay in Minutes']=df1['Arrival Delay in Minutes'].fillna(df1['Arrival Delay in Minutes'].median())
```

```python
df1.isnull().sum()
```

#### Now we need to change the categorical features of some columns to numerical features for ease of our model

We will use label encoding for "Satisfaction" and "Gender" as they have only two possible values

```python
print(df1['satisfaction'].unique())
print(df1['Gender'].unique())
```

```python
df1['Gender']=df1['Gender'].map({'Male': 1, 'Female': 0})
```
