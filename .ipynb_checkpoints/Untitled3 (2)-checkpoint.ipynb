```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,classification_report
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('/train.csv')
print(df.shape)
```

```python
test_df = pd.read_csv('/test.csv')
print(test_df.shape)
```

```python
for col in df.columns:
    print(col)
```

```python
ln = LabelEncoder()

categorical_columns = ['satisfaction', 'Gender', 'Customer Type', 'Type of Travel', 'Class']
for col in categorical_columns:
    df[col] = ln.fit_transform(df[col])

categorical_columns = ['Gender', 'Customer Type', 'Type of Travel', 'Class']
for col in categorical_columns:
    test_df[col] = ln.fit_transform(test_df[col])
```

```python
null_values = df.isnull().sum()
print(null_values)
```

```python
null_values = test_df.isnull().sum()
print(null_values)
```

```python
correlation_matrix = df.drop(columns=["Id"]).corr()


plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap ')
plt.show()
```

```python
df.hist(figsize=(20,20))
```

```python
var=df.drop(['Id','satisfaction','Gender', 'Class','Arrival Delay in Minutes'], axis=1)
```

```python
target=df[['satisfaction']]
```

```python
test_df_pred=test_df.drop(['Id','Gender', 'Class','Arrival Delay in Minutes'], axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(var,target,test_size=0.2,random_state=42)
```

```python
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy}')
```

```python
def createSubmission(test_df , y_pred, filename):
    y_pred = model.predict(test_df_pred)
    y_pred = ['unsatisfied' if pred == 0 else 'satisfied' for pred in y_pred]
    submission = pd.DataFrame({'ID': test_df['Id'] , 'satisfaction': y_pred})
    submission.to_csv(filename , index=False)
```

```python
filename="Banished_Parlor_7_10_last.csv"
```

```python
createSubmission(test_df, y_pred,filename)
```
